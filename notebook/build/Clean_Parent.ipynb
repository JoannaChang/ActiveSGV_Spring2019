```python
import pandas as pd
df = pd.read_csv(clean_path + "Spanish_and_English_Parent.csv", encoding = 'latin1')
df
```

```python
df = df.drop(["Collector ID", "Start Date", "End Date", "IP Address", "Email Address", "First Name", "Last Name", "Custom Data 1" ], axis = 1)
df.head()
```

```python
# df.colums = [col.strip() for col in df.columns]
# df.head()

df = df.rename(str.strip, axis=1)
df.columns
```

```python
# Given a column (as an improvement, as a concern), reformat values to be T / F
def col_to_bool(col_name):
    return pd.isna(df[col_name]) == False
   
```

```python
def insert_col_values(insert_into, insert_from, keyword):
    df[df[insert_into] == keyword][insert_into] = df[df[insert_into] == keyword][insert_from]
```

```python
as_concern = [c for c in df.columns if c.lower()[-13:] == ' as a concern']
df[as_concern].columns
```

```python
for c in df[as_concern].columns:
    df[c] = col_to_bool(c)
df[as_concern].head()
```

```python
as_improvement = [c for c in df.columns if c.lower()[-18:] == ' as an improvement']
df[as_improvement].columns
```

```python
# convert to bool values
for c in df[as_improvement].columns:
    df[c] = col_to_bool(c)
df[as_improvement].head()
```

```python
# lower case all the street names
df['What is the street intersection nearest your home? (Provide the names of two intersecting streets)'] = df['What is the street intersection nearest your home? (Provide the names of two intersecting streets)'].str.lower()
df.head(2)
```

```python
how_far_translation = {'Menos de 1/4 de milla': 'Less than 1/4 of a mile', 
                       '1/4-1/2 milla': '1/4 to 1/2 of a mile', 
                      '1/2-1 milla': '1/2-1 mile',
                      '1-2 milla': '1-2 miles',
                      'No se': 'I do not know',
                      '2+ millas': '2+ miles'}
```

```python
to_from_transpo_translation = {'Vehiculo familiar (solo con ninos de la familia)': 'Family Vehicle (only children in your family)',
                                'Caminando': 'Walking',
                                'Autobus escolar': 'School Bus',
                                'Carpool (auto compartido con ninos de otras familias)': 'Carpool (children from other families)',
                                }
```

```python
time_translation = {'Menos de 5 minutos': 'Less than 5 minutes',
                    '5 to 10 minutos': '5-10 minutes', 
                    '11 to 20 minutos': '11-20 minutes',
                    'Mas de 20 minutos': 'More than 20 minutes',
                    'No se/No estoy seguro/a': 'I do not know'}
```

```python
at_what_grade_translation = {'No me sentiria comodo/a en ningun grado': -1}
```

```python
# func to translate things
def translate_to_eng(column_name, dict):
    df[column_name] = df[column_name].map(lambda val: dict[val] if val in dict else val)
```

```python
# translating
translate_to_eng('How far does your child live from school?', how_far_translation)
translate_to_eng('On most days, how does your child get to school?', to_school_transpo_translation)
translate_to_eng('On most days, how does your child leave school?', to_school_transpo_translation)
translate_to_eng('How long does it normally take your child to get to school?', time_translation)
translate_to_eng('How long does it normally take your child to get home from school?', time_translation)
translate_to_eng('At what grade would you allow your child to walk or bike to/from school without an adult?', at_what_grade_translation)
df
```

```python
def replace_value(col, df, content, want):
    df[col] = df[col].str.replace(content, want)
    return df

df = replace_value('At what grade would you allow your child to walk or bike to/from school without an adult?', 
              df, "th", "")
df = replace_value('At what grade would you allow your child to walk or bike to/from school without an adult?', 
              df, "th", "")
```
